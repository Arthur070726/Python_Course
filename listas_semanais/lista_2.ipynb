```
l = []
def quadrado_par(N):
    for x in range(N):
        if x%2 == 0:
            l.append(x **2)
    return l
```

problema 7

```
def intercept(A , B):
   set_c = set(B).intersection(set(A))
   l = list(set_c)
   l.sort()
   return l
```

```
A = [1 , 2 , 3 , 4 , 5, 6]
B = [1 , 4 , 7, 5 , 7 , 8]
intercept(A , B)
```

problema 8

```
def union(A , B):
   set_c = set(B).union(set(A))
   l = list(set_c)
   l.sort()
   return l
```

```
A = [-2, 0, 1, 2]
B = [-1, 1, 2, 10]
union(A,B)
```
